```python
#Trial Update Frequency Quantiles: What does the distribution of trial update frequencies look like across four quantiles?
```

```python
    STEP 1 # Convert "Last Update Posted" and "First Posted" to datetime format
```

```python
#Trial Update Frequency Quantiles: What does the distribution of trial update frequencies look like across four quantiles?
```

```python
df['First Posted'] = pd.to_datetime(df['First Posted'])
df['Last Update Posted'] = pd.to_datetime(df['Last Update Posted'])

df['Update Frequency'] = (df['Last Update Posted'] - df['First Posted']).dt.days
```

```python
         #make bins using qcut
```

```python
q = 5
cut_labels = ['1st Quantile', '2nd Quantile', '3rd Quantile', '4th Quantile']

df['UpdateFrequency_Q'] = pd.qcut(df['Update Frequency'], 5, labels = cut_labels, duplicates='drop')
```

```python
        # sorting of bins
```

```python
df['UpdateFrequency_Q'].value_counts().sort_index()
```

```python
            GRAPHICALLY REPRESENTATION
```

```python
df['UpdateFrequency_Q'].value_counts().sort_index().plot(kind = 'bar', color = 'purple')
plt.xlabel( 'days')
plt.ylabel('number of days')
plt.title('Distribution of trials')
plt.show()
```

```python
#Completion Time Frame Grouping: Apply the cut function to categorize the completion time frames of trials into distinct intervals. What are the characteristics of trials within each time frame group?
```

```python
                #finding number of days
```

```python
df['Completion Time Frame (Days)'] = (df['Completion Date'] - df['Start Date']).dt.days
```

```python
               #making bins
```

```python
cut_bins_2 = [0, 180, 365, 730, 1460, np.inf] #Total days
labels_2 = ['0-6 months', '6-12 months', '1-2 years', '2-4 years', '> 4 years']

df['CompletionTimeFrame_Q'] = pd.cut(df['Completion Time Frame (Days)'], bins=cut_bins_2, labels=labels_2)
df['CompletionTimeFrame_Q'].value_counts().sort_index()
```
