**5. Solve following question by using qcut**

```python
# Enrollment Quantile Categories: How are trials distributed across four quantiles based on enrollment numbers?

df['Enrollments_Q'] = pd.qcut(df['Enrollment'],q=4)
print(df[['Enrollment','Enrollments_Q']])
```

```python
df['Enrollments_Q'].value_counts().sort_index()
```

```python
# Start Date Year Quantiles: What are the time periods represented by four quantiles of trial start years?

df['Time_Q'] = pd.qcut(df['Start Date'],q=4)
print(df[['Start Date','Time_Q']])
```

```python
df['Time_Q'].value_counts().sort_index()
```

```python
# Completion Time Frame Categories: How do trial durations distribute across four quantiles from start to completion date?

df['Completion Date'] = pd.to_datetime(df['Completion Date'], format='mixed')

df['Completion_Q'] = pd.qcut(df['Completion Date'],q=4)
print(df[['Completion Date','Completion_Q']])
```

```python
df['Completion_Q'].value_counts().sort_index()
```

```python
# Outcome Measures Count Quantiles: How are trials categorized into four quantiles based on the number of outcome measures?

df['OutcomeMeasures_Count'] = df['Outcome Measures'].str.count('\|') + 1 # Outcome measures separated by a character (e.g., "|")

df['OutcomeMeasures_Q'] = pd.qcut(df['OutcomeMeasures_Count'], 4, labels=['Low', 'Medium', 'High', 'Extreme'])
print(df[['Outcome Measures','OutcomeMeasures_Q']])
```

```python
df['OutcomeMeasures_Q'].value_counts().sort_index()
```

```python
# Trial Update Frequency Quantiles: What does the distribution of trial update frequencies look like across four quantiles?

df['First Posted'] = pd.to_datetime(df['First Posted'])
df['Last Update Posted'] = pd.to_datetime(df['Last Update Posted'])

df['Update Frequency'] = (df['Last Update Posted'] - df['First Posted']).dt.days
```

```python
cut_bins = 5 #Bin labels must be one fewer than the number of bin edges
cut_labels = ['1st Quantile', '2nd Quantile', '3rd Quantile', '4th Quantile']

df['UpdateFrequency_Q'] = pd.qcut(df['Update Frequency'], cut_bins, labels = cut_labels, duplicates='drop')
```

```python
df['UpdateFrequency_Q'].value_counts().sort_index()
```

**6. Completion Time Frame Grouping: Apply the cut function to categorize the completion time frames of trials into distinct intervals. What are the characteristics of trials within each time frame group?**

```python
# Convert "Start Date" and "Completion Date" to datetime format
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['Completion Date'] = pd.to_datetime(df['Completion Date'])

# Calculate the completion time frame as the number of days between the two dates
df['Completion Time Frame (Days)'] = (df['Completion Date'] - df['Start Date']).dt.days
```

```python
_bins = [0, 180, 365, 730, 1460, np.inf] #Total days
_labels = ['0-6 months', '6-12 months', '1-2 years', '2-4 years', '> 4 years']

df['CompletionTimeFrame_Q'] = pd.cut(df['Completion Time Frame (Days)'], bins=_bins, labels=_labels)
df['CompletionTimeFrame_Q'].value_counts().sort_index()
```
